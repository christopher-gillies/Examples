```python
from scipy import signal
import matplotlib.pyplot as plt
import numpy as np
import scipy.io.wavfile as wav
import IPython.display as ipd
```

```python
sr = 22050 # sample rate
T = 1.0    # seconds
t = np.linspace(0, T, int(T*sr), endpoint=False)
print(t.shape)
# Duty is the amount of time the waveform is on
square_wave = 0.5 * signal.square(2 * np.pi * 500 * t,duty=0.5)

plt.plot(t, signal.square(2 * np.pi * 5 * t,duty=0.5))
plt.ylim(-1, 1)
```

```python
ipd.Audio(data=square_wave, rate=sr) # load a NumPy array
```

```python
saw_tooth = signal.sawtooth(2 * np.pi * 500 * t)
plt.plot(t, signal.sawtooth(2 * np.pi * 5 * t))
ipd.Audio(data=saw_tooth, rate=sr)   
```
