```python
from scipy import signal
import matplotlib.pyplot as plt
import numpy as np
import scipy.io.wavfile as wav
import IPython.display as ipd
```

```python
t = np.linspace(0, 1, 1000, endpoint=False)
print(t.shape)
square_wave = signal.square(2 * np.pi * 100 * t,duty=0.5)
plt.plot(t, square_wave)
plt.ylim(-2, 2)
```

```python
wav.write("/Users/cgillies/Downloads/test.wav", rate=1000, data=square_wave)
ipd.Audio(data=square_wave, rate=1000) # load a NumPy array
```
